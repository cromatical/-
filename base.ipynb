# using for

```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
import time

def use_for(r):
    center_x = r
    center_y = r
    x = np.arange(0, 2*r)
    y = np.arange(0, 2*r)

    square_yellow = np.ones((2*r, 2*r, 3))

    square_yellow[:,:,2] = 0

    for i in x:
        for j in y:
            if ((i-center_x)**2 + (j-center_y)**2 < r**2):
                square_yellow[i,j,0:2]=0
                square_yellow[i,j,2]=1
    return square_yellow

def no_for(r):
    center_x = r
    center_y = r
    x = np.arange(0, 2*r)
    y = np.arange(0, 2*r)

    square_yellow = np.zeros((2*r, 2*r))
    square_yellow_x=square_yellow+x
    x1 = (square_yellow_x - r) ** 2
    square_yellow_y=np.transpose(square_yellow+y)
    y1 = (square_yellow_y - r) **2

    circle = x1+y1
    circle1 = np.where(circle < r**2, 1, 0)

    square_yellow = np.zeros((2*r, 2*r, 3))
    square_yellow[:,:,[0,1]] = 1
    square_yellow[:,:,2] = square_yellow[:,:,2] + circle1
    square_yellow[:,:,1] = square_yellow[:,:,1] - circle1
    square_yellow[:,:,0] = square_yellow[:,:,0] - circle1
    return square_yellow


if __name__="__main__":
    
    r = [100, 200, 300, 400, 500, 1000, 2000, 3000, 4000, 5000]
    y = []
    y1 = []

    for i in r:
        start = time.time()
        use_for(i)
        end = time.time()
        result1 = end - start
        y.append(result1)

        start1 = time.time()
        no_for(i)
        end1 = time.time()
        result2 = end1 - start1
        y1.append(result2)
```

```python
print(y)
print(y1)
```

```python
import matplotlib.pyplot as plt

plt.plot(r, y, label = 'use_for', marker = 'o')
plt.plot(r, y1, label = 'no_for', marker = 'o')

plt.xlabel('r')
plt.ylabel('time(sec)')

plt.legend()
plt.show()
```
